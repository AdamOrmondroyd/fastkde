```python
import sys
sys.path.append("../")
import selfConsistentDensityEstimate as sc
import numpy as np
%load_ext autoreload
%autoreload 2

N = 2e5
var1 = 50*np.random.normal(size=N) + 0.1
var2 = 0.01*np.random.normal(size=N) - 300

pdf,axis = sc.pdf(var1)
```

```python
%matplotlib inline
import pylab as PP

PP.semilogy(axis,pdf)
PP.show()
```

```python
pdf2,axes = sc.pdf(var1=var1,var2=var2)
PP.contourf(axes[0],axes[1],pdf2,128)
PP.show()
```

```python
def zfunc(delta,f,dx,floor):
    ipos = np.where((f-delta) >= floor)
    return 1. - sum((f[ipos]-delta)*dx)
```

```python
from scipy.optimize import brentq
```

```python
dx = axis[1]-axis[0]
a =  zfunc(0,pdf,dx,(1./len(var1)))
b = -1e-2
delta = brentq(zfunc,a,b,(pdf,dx,1000./len(var1)),xtol=1e-17)
```

```python
pp = pdf - delta
pp[np.where(pp < 0)] = 0.0
```

```python
PP.contour(axes[0],axes[1],pdf2)
PP.show()
```
