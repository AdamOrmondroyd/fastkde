```python
%load_ext autoreload
%autoreload 2
from nufft import idft
from nufft import nuifft as nuFFT
import pylab as PP
import numpy as np
import numpy.fft as fft
import matplotlib.colors as col
%matplotlib inline
```

```python

x = np.linspace(-2*np.pi,2*np.pi,65)
y = np.linspace(-4*np.pi,4*np.pi,33)
xx,yy = np.meshgrid(x,y)
zz = np.sin(2*xx)*np.cos(4*yy)
ifilterx = np.where(abs(xx) > np.pi)
ifiltery = np.where(abs(yy) > 2*np.pi)
zz[ifilterx] = 0.0
zz[ifiltery] = 0.0


PP.imshow(zz)
PP.show()
```

```python
def calcXfromT(tpoints):
  """Calculates real space points given a set of hermetian frequency points. """
  #Use fftfreq to produce a set of frequencies that correspond to the fourier
  #transform of a signal on the tpoints points
  deltaT = tpoints[1] - tpoints[0]
  return  fft.fftshift(fft.fftfreq(len(tpoints),deltaT/(2*np.pi)))

def calcTfromX(xpoints):
  
  """Calculates frequency points given a signal in real space. """
  #Use fftfreq to produce a set of frequencies that correspond to the fourier
  #transform of a signal on the x points
  deltaX = xpoints[1] - xpoints[0]
  return fft.fftshift(fft.fftfreq(len(xpoints),deltaX/(2*np.pi)))
```

```python
tx = calcXfromT(x)
ty = calcXfromT(y)

fftzz = fft.fftshift(fft.ifftn(zz))
```

```python
PP.contourf(tx,ty,abs(fftzz))
PP.colorbar()
PP.show()
```

```python
ntmax = max(len(tx),len(ty))
fillVal = -1e20
txty = fillVal*np.ones([2,ntmax])
txty[0,:len(tx)] = tx
txty[1,:len(ty)] = ty
xy = np.concatenate((xx.ravel()[np.newaxis,:],yy.ravel()[np.newaxis,:]))

dftzz = idft(xy,np.array(zz.ravel(),dtype=np.complex128),txty,missingFreqVal=fillVal)
```

```python
PP.contourf(tx,ty,abs(dftzz).T)
PP.colorbar()
PP.show()
```

```python
nufftzz = nuFFT(xy,np.array(zz.ravel(),dtype=np.complex128),txty,missingFreqVal=fillVal,beVerbose=1)
```

```python
PP.contourf(tx,ty,abs(nufftzz).T)
PP.colorbar()
PP.show()
```
